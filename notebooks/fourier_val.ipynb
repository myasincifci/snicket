```python
import os
import random
from PIL import Image
import numpy as np

import matplotlib.pyplot as plt
```

```python
# Directory containing images
image_dir_midj = '../data/val/midjourney'
image_dir_nat = '../data/val/nature'

# Get list of all files in the directory
all_files_midj = os.listdir(image_dir_midj)
all_files_nat = os.listdir(image_dir_nat)

# # Filter out non-image files
# image_files = [f for f in all_files if f.endswith(('png', 'jpg', 'jpeg', 'bmp', 'gif'))]

# Select a random image file
random_image_file_midj = random.choice(all_files_midj)
random_image_file_nat = random.choice(all_files_nat)

# Load the image
image_path_midj = os.path.join(image_dir_midj, random_image_file_midj)
image_path_nat = os.path.join(image_dir_nat, random_image_file_nat)
image_midj = Image.open(image_path_midj)
image_nat = Image.open(image_path_nat)  

# Display the image
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the midjourney image
axes[0].imshow(image_midj)
axes[0].set_title('Midjourney Image')
axes[0].axis('off')  # Hide axes

# Plot the nature image
axes[1].imshow(image_nat)
axes[1].set_title('Nature Image')
axes[1].axis('off')  # Hide axes

plt.show()
```

```python
# Convert the midjourney image to grayscale
image_midj_gray = image_midj.convert('L')

# Convert the grayscale midjourney image to a numpy array
image_midj_array = np.array(image_midj_gray)

# Perform FFT on the midjourney image
fft_result_midj = np.fft.fft2(image_midj_array)

# Shift the zero-frequency component to the center of the spectrum for midjourney image
fft_shifted_midj = np.fft.fftshift(fft_result_midj)

# Get the magnitude spectrum for midjourney image
magnitude_spectrum_midj = 20 * np.log(np.abs(fft_shifted_midj))

# Convert the nature image to grayscale
image_nat_gray = image_nat.convert('L')

# Convert the grayscale nature image to a numpy array
image_nat_array = np.array(image_nat_gray)

# Perform FFT on the nature image
fft_result_nat = np.fft.fft2(image_nat_array)

# Shift the zero-frequency component to the center of the spectrum for nature image
fft_shifted_nat = np.fft.fftshift(fft_result_nat)

# Get the magnitude spectrum for nature image
magnitude_spectrum_nat = 20 * np.log(np.abs(fft_shifted_nat))

# Plot the magnitude spectra
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot the magnitude spectrum for midjourney image
axes[0].imshow(magnitude_spectrum_midj, cmap='gray')
axes[0].set_title('Magnitude Spectrum (Midjourney)')
axes[0].axis('off')  # Hide axes

# Plot the magnitude spectrum for nature image
axes[1].imshow(magnitude_spectrum_nat, cmap='gray')
axes[1].set_title('Magnitude Spectrum (Nature)')
axes[1].axis('off')  # Hide axes

plt.show()
```

```python
# Directory to save Fourier transform images
output_dir_midj = '../data/val/midjourney_ft'
os.makedirs(output_dir_midj, exist_ok=True)

# Iterate through each image file in the midjourney directory
for image_file in all_files_midj:
    # Load the image
    image_path = os.path.join(image_dir_midj, image_file)
    image = Image.open(image_path)
    
    # Convert the image to grayscale
    image_gray = image.convert('L')
    
    # Convert the grayscale image to a numpy array
    image_array = np.array(image_gray)
    
    # Perform FFT on the image
    fft_result = np.fft.fft2(image_array)
    
    # Shift the zero-frequency component to the center of the spectrum
    fft_shifted = np.fft.fftshift(fft_result)
    
    # Get the magnitude spectrum
    magnitude_spectrum = 20 * np.log(np.abs(fft_shifted))
    
    # Save the magnitude spectrum as an image
    output_path = os.path.join(output_dir_midj, image_file)
    plt.imsave(output_path, magnitude_spectrum, cmap='gray')
```

```python
# Directory to save Fourier transform images
output_dir_nat = '../data/val/nature_ft'
os.makedirs(output_dir_nat, exist_ok=True)

# Iterate through each image file in the midjourney directory
for image_file in all_files_nat:
    # Load the image
    image_path = os.path.join(image_dir_nat, image_file)
    image = Image.open(image_path)
    
    # Convert the image to grayscale
    image_gray = image.convert('L')
    
    # Convert the grayscale image to a numpy array
    image_array = np.array(image_gray)
    
    # Perform FFT on the image
    fft_result = np.fft.fft2(image_array)
    
    # Shift the zero-frequency component to the center of the spectrum
    fft_shifted = np.fft.fftshift(fft_result)
    
    # Get the magnitude spectrum
    magnitude_spectrum = 20 * np.log(np.abs(fft_shifted))
    
    # Save the magnitude spectrum as an image
    output_path = os.path.join(output_dir_nat, image_file)
    plt.imsave(output_path, magnitude_spectrum, cmap='gray')
```
